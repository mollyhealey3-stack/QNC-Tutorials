This figure is from a paper measuring saccade latency in trials rewarded for rightward saccades, it illustrates a gaussian distribution. https://doi.org/10.1038/nature00892

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated data resembling the figure (centered around 220 ms, narrow spread)
np.random.seed(0)
saccade_latencies = np.random.normal(loc=220, scale=20, size=500)
saccade_latencies = saccade_latencies[(saccade_latencies > 100) & (saccade_latencies < 350)]

# making the histogram (parameters: size, axes labels, color, etc.)
plt.figure(figsize=(4, 4))  # Square figure
plt.hist(saccade_latencies, bins=30, color='navy')

plt.title('Reward', pad=10)
plt.xlabel('Saccade latency (ms)')
plt.ylabel('Number of trials')
plt.xlim(0, 600)
plt.ylim(0, 100)
plt.tight_layout()

# Save or show
plt.show()
```
